```python
#Loading the Dataset
df = pd.read_csv(working_dir_path + 'assignment1.csv')
```

```python
df['date'] = pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
import numpy as np

# Create a dictionary to store the previous attempts for each student and subject
previous_attempts = {}

# Iterate over each row and fill in the marks for each attempt
for index, row in df.iterrows():
    student = row['student_roll_no']
    subject = row['subject']
    
    # Get the previous attempts for this student and subject
    if (student, subject) not in previous_attempts:
        previous_attempts[(student, subject)] = []
    attempt_marks = [np.nan] * len(previous_attempts[(student, subject)]) + [row['marks']]
    previous_attempts[(student, subject)].append(attempt_marks)
    
    # Fill in the marks for each attempt
    for i, attempt_marks in enumerate(previous_attempts[(student, subject)][::-1], start=1):
        df.at[index, f'm{i}'] = attempt_marks[-1]

# Fill in NaN for any missing marks
df.fillna(value=0, inplace=True)
```
